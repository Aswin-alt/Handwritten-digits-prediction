```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
data = pd.read_csv("car.data")
print(data.head())
```

```python
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
air = le.fit_transform(list(data["air"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
```

```python
print(buying)
print(air)
print(safety)
print(cls)
```

```python
predict = "class"

X = list(zip(buying, maint, air, persons, lug_boot, safety))
y = list(cls)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
n_neighbors=7
model = KNeighborsClassifier(n_neighbors)
```

```python
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc*100)
```

```python
predicted = model.predict(x_test)
names = ["poor", "average", "good", "vgood"]
```

```python
for x in range(len(predicted)):
    print("Predicted:", names[predicted[x]], "\tData:", x_test[x], "\tActual:", names[y_test[x]])
    n = model.kneighbors([x_test[x]], n_neighbors, True)
    print("N: ", n)
```
